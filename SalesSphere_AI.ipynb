MODEL TRAINING

```python
X = df[['Age' , 'Gender','City','Income','Membership_Level','Product_Category','Product_Name','Base_Price','Discount_Percentage','Season','Payment_Method']].copy()
y = df['Buy_or_Not']
```

```python
categorical_cols = [
    'Gender',
    'City',
    'Membership_Level',
    'Product_Category',
    'Product_Name',
    'Season',
    'Payment_Method'
]

X = pd.get_dummies(
    X,
    columns=categorical_cols,
    drop_first=True
)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

Model Training


1)Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr_model = LogisticRegression(max_iter=1000,solver='liblinear')
```

```python
lr_model.fit(X_train,y_train)
```

```python
y_pred = lr_model.predict(X_test)
```

Accuracy for LogisticRegression model

```python
from sklearn.metrics import accuracy_score
```

```python
lr_accuracy = accuracy_score(y_test,y_pred)
lr_accuracy
```

Confusion Matrix for LogisticRegression

```python
from sklearn.metrics import confusion_matrix
```

```python
logistic_cm = confusion_matrix(y_test,y_pred)
logistic_cm
```

```python
sns.heatmap(logistic_cm,annot=True,fmt='d')
plt.title('Logistic Regression Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

Simulation

```python
simulation_df = df.copy()
```

```python
simulation_df['Product_Category'] = 'Personal Care'
simulation_df['Product_Name'] = 'Shampoo'
simulation_df['Base_Price'] = 180
simulation_df['Discount_Percentage'] = 15
simulation_df['Season'] = 'Summer'
```

```python
simulation_X = simulation_df[
    [
        'Age',
        'Gender',
        'City',
        'Income',
        'Membership_Level',
        'Product_Category',
        'Product_Name',
        'Base_Price',
        'Discount_Percentage',
        'Season',
        'Payment_Method'
    ]
].copy()
```

```python
simulation_X = pd.get_dummies(
    simulation_X,
    columns=[
        'Gender',
        'City',
        'Membership_Level',
        'Product_Category',
        'Product_Name',
        'Season',
        'Payment_Method'
    ],
    drop_first=True
)
```

```python
simulation_X = simulation_X.reindex(
    columns=X.columns,
    fill_value=0
)
```

```python
probabilities = xgb_tuned.predict_proba(simulation_X)[:,1]
```

```python
avg_probability = probabilities.mean() * 100
```

```python
buyers = (probabilities >= 0.5).sum()
```

```python
total_customers = len(probabilities)

demand_percentage = (
    buyers / total_customers
) * 100
```

```python
print("Total Customers:", total_customers)
print("Potential Buyers:", buyers)
print("Demand Percentage:", round(demand_percentage,2))
print("Average Purchase Probability:", round(avg_probability,2))
```
